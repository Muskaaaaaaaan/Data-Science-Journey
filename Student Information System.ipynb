# Make a Mini Project

## Student Information System

#### The program should:

##### Ask for:

Name, Age, Course, Marks in 3 subjects, Calculate: , Total, Average, Display the information in a clean format.

```python
# Student Information System
print("========== Student Information System ==========")
# Student Details
name = input("Enter Student Name: ")
age = int(input("Enter Age: "))
course = input("Enter Course: ")

# Marks
mark1 = float(input("Enter Marks of Subject 1: "))
mark2 = float(input("Enter Marks of Subject 2: "))
mark3 = float(input("Enter Marks of Subject 3: "))

# Calculations
total = mark1 + mark2 + mark3
average = total / 3

# Output
print("\n========== Student Report ==========")
print("Name: ", name)
print("Age: ", age)
print("Course:", course)
print("Subject 1 : ", mark1)
print("Subject 2 : ", mark2)
print("Subject 3 : ", mark3)
print("------------------------------------")
print("Total: ", total)
print("Average: ", average)
print("====================================")
```
