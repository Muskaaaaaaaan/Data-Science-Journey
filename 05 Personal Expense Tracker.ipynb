## Personal Expense Tracker
Features: Add income, Add expense, View balance, View all transactions, Search transactions, Delete a transaction, Exit

Skills Used: Functions, Strings, Lists, Loops, Conditionals, Modular programming

```python
transactions = []
def add_income():
    amount = float(input("Enter income amount: "))
    transactions.append(["Income", amount])
    print("Income Added Successfully!")
def add_expense():
    amount = float(input("Enter expense amount: "))
    transactions.append(["Expense", amount])
    print("Expense Added Successfully!")
def view_balance():
    income = 0
    expense = 0
    for item in transactions:
        if item[0] == "Income":
            income = income + item[1]
        else:
            expense = expense + item[1]
    balance = income - expense
    print("Total Income :", income)
    print("Total Expense:", expense)
    print("Balance       :", balance)
def view_transactions():
    if len(transactions) == 0:
        print("No Transactions Found.")
    else:
        print("\n------ Transactions ------")
        for i in range(len(transactions)):
            print(i + 1, ".", transactions[i][0], "-", transactions[i][1])
def search_transaction():
    word = input("Enter Income or Expense: ")
    found = False
    for item in transactions:
        if item[0].lower() == word.lower():
            print(item[0], "-", item[1])
            found = True
    if found == False:
        print("Transaction Not Found.")
def delete_transaction():
    view_transactions()
    if len(transactions) == 0:
        return
    num = int(input("Enter transaction number to delete: "))
    if num >= 1 and num <= len(transactions):
        transactions.pop(num - 1)
        print("Transaction Deleted.")
    else:
        print("Invalid Transaction Number.")
while True:
    print("\n========== PERSONAL EXPENSE TRACKER ==========")
    print("1. Add Income")
    print("2. Add Expense")
    print("3. View Balance")
    print("4. View All Transactions")
    print("5. Search Transactions")
    print("6. Delete Transaction")
    print("7. Exit")
    choice = int(input("Enter your choice: "))
    if choice == 1:
        add_income()
    elif choice == 2:
        add_expense()
    elif choice == 3:
        view_balance()
    elif choice == 4:
        view_transactions()
    elif choice == 5:
        search_transaction()
    elif choice == 6:
        delete_transaction()
    elif choice == 7:
        print("Thank You!")
        break
    else:
        print("Invalid Choice.")
```
