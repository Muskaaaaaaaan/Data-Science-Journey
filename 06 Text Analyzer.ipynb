# Text Analyzer

Input:

Enter a paragraph:

Output: Total characters, Total words, Total sentences, Longest word, Shortest word, Most common word, Number of vowels, Number of consonants, Reading time estimate, Average word length

```python
def text_analyzer():
    text = input("Enter a paragraph:\n")
# Total Characters
    total_characters = len(text)
# Total Words
    words = text.split()
    total_words = len(words)
# Total Sentences
    sentences = 0
    for ch in text:
        if ch == "." or ch == "!" or ch == "?":
            sentences = sentences + 1
    if sentences == 0:
        sentences = 1
# Longest Word
    longest = words[0]
    for word in words:
        if len(word) > len(longest):
            longest = word
# Shortest Word
    shortest = words[0]
    for word in words:
        if len(word) < len(shortest):
            shortest = word# Vowels and Consonants
    vowels = 0
    consonants = 0
    for ch in text.lower():
        if ch.isalpha():
            if ch == "a" or ch == "e" or ch == "i" or ch == "o" or ch == "u":
                vowels = vowels + 1
            else:
                consonants = consonants + 1
# Most Common Word
    frequency = {}
    for word in words:
        word = word.lower()
        if word in frequency:
            frequency[word] = frequency[word] + 1
        else:
            frequency[word] = 1
    most_common = ""
    highest = 0
    for word in frequency:
        if frequency[word] > highest:
            highest = frequency[word]
            most_common = word
# Reading Time
    reading_time = total_words / 200
# Average Word Length
    total_letters = 0
    for word in words:
        total_letters = total_letters + len(word)
    average_length = total_letters / total_words
    print("\n========== TEXT ANALYSIS ==========")
    print("Total Characters :", total_characters)
    print("Total Words      :", total_words)
    print("Total Sentences  :", sentences)
    print("Longest Word     :", longest)
    print("Shortest Word    :", shortest)
    print("Most Common Word :", most_common)
    print("Vowels           :", vowels)
    print("Consonants       :", consonants)
    print("Reading Time     :", round(reading_time, 2), "minutes")
    print("Average Length   :", round(average_length, 2))
text_analyzer()
```
