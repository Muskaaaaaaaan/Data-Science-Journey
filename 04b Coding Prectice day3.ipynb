### Functions

```python
# Function to add two numbers.
def add(a, b):
    return a + b
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

result = add(num1, num2)

print("Sum =", result)
```

```python
# Function to calculate the square of a number.
def square(num):
    return num * num
n = int(input("Enter a number: "))
print("Square =", square(n))
```

```python
# Function to check even or odd.
def check(num):
    if num % 2 == 0:
        print("Even Number")
    else:
        print("Odd Number")
n = int(input("Enter a number: "))
check(n)
```

```python
# Function to find the largest of three numbers.
def largest(a, b, c):
    if a > b and a > c:
        return a
    elif b > a and b > c:
        return b
    else:
        return c
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))
print("Largest =", largest(num1, num2, num3))
```

```python
# Function to calculate factorial.
def factorial(num):
    fact = 1
    for i in range(1, num + 1):
        fact = fact * i
    return fact
n = int(input("Enter a number: "))
print("Factorial =", factorial(n))
```

```python
# Function to check prime numbers.
def prime(num):
    count = 0
    for i in range(1, num + 1):
        if num % i == 0:
            count = count + 1
    if count == 2:
        print("Prime Number")
    else:
        print("Not a Prime Number")
n = int(input("Enter a number: "))
prime(n)
```

```python
# Function to calculate average marks.
def average(m1, m2, m3):
    avg = (m1 + m2 + m3) / 3
    return avg
mark1 = int(input("Enter first mark: "))
mark2 = int(input("Enter second mark: "))
mark3 = int(input("Enter third mark: "))
print("Average =", average(mark1, mark2, mark3))
```

```python
# Function using default arguments.
def greet(name="Student"):
    print("Welcome", name)
greet()
greet("Sanjana")
```

```python
# Function using keyword arguments.
def student(name, age):
    print("Name:", name)
    print("Age:", age)
student(age=24, name="Sanjana")
```

```python
# Lambda function for multiplication.
multiply = lambda a, b: a * b
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
print("Multiplication =", multiply(num1, num2))
```

```python
# Mini Practice Project – Student Result Calculator
def average(m1, m2, m3):
    return (m1 + m2 + m3) / 3
mark1 = int(input("Enter English marks: "))
mark2 = int(input("Enter Maths marks: "))
mark3 = int(input("Enter Science marks: "))
avg = average(mark1, mark2, mark3)
print("Average =", avg)
if avg >= 40:
    print("Result: Pass")
else:
    print("Result: Fail")
```

### Strings

```python
# Reverse a string.
text = input("Enter a string: ")
reverse = " "
for i in text:
    reverse = i + reverse
print("Reversed String =", reverse)
```

```python
# Check if a string is a palindrome.
text = input("Enter a string: ")
reverse = ""
for i in text:
    reverse = i + reverse
print("Original =", text)
print("Reverse =", reverse)
if text == reverse:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

```python
# Count vowels and consonants.
text = input("Enter a string: ")
vowel = 0
consonant = 0
for ch in text:
    if ch.isalpha():
        if ch == 'a' or ch == 'e' or ch == 'i' or ch == 'o' or ch == 'u' or \
           ch == 'A' or ch == 'E' or ch == 'I' or ch == 'O' or ch == 'U':
            vowel = vowel + 1
        else:
            consonant = consonant + 1
print("Vowels =", vowel)
print("Consonants =", consonant)
```

```python
# Count words in a sentence.
sentence = input("Enter a sentence: ")
words = sentence.split()
print("Total Words =", len(words))
```

```python
# Find character frequency.
text = input("Enter a string: ")
ch = input("Enter a character: ")
count = 0
for i in text:
    if i == ch:
        count = count + 1
print("Frequency =", count)
```

```python
# Replace spaces with underscores.
text = input("Enter a sentence: ")
new_text = ""
for ch in text:
    if ch == " ":
        new_text = new_text + "_"
    else:
        new_text = new_text + ch
print(new_text)
```

```python
# Check if a string starts with a given word.
text = input("Enter a sentence: ")
word = input("Enter the starting word: ")
if text.startswith(word):
    print("Yes, it starts with", word)
else:
    print("No, it does not start with", word)
```

```python
# Convert a sentence to title case.
text = input("Enter a sentence: ")
print(text.title())
```

```python
# Remove duplicate characters.
text = input("Enter a string: ")
result = ""
for ch in text:
    if ch not in result:
        result = result + ch
print("After Removing Duplicates =", result)
```

```python
# Find the longest word in a sentence.
sentence = input("Enter a sentence: ")
words = sentence.split()
longest = words[0]
for word in words:
    if len(word) > len(longest):
        longest = word
print("Longest Word =", longest)
```

```python
# Mini Project – Password Strength Checker
password = input("Enter your password: ")
if len(password) < 8:
    print("Weak Password")
else:
    has_digit = False
    for ch in password:
        if ch.isdigit():
            has_digit = True
    if has_digit:
        print("Strong Password")
    else:
        print("Medium Password")
```
